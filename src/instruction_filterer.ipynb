```python
import json
import os

def contains_keyword(sentence, keywords):
    sentence_lower = sentence.lower()
    keywords_lower = [keyword.lower() for keyword in keywords]

    for keyword in keywords_lower:
        if keyword in sentence_lower:
            return True
    return False

filter_words_path = '/content/drive/MyDrive/Project/filter_words.json'
########################
original_combined_path = '/content/drive/MyDrive/Project/finetune data/combined_output.json'
######################## Edited by 光齊
# combined_path = '/content/drive/MyDrive/Project/finetune data/short_combined_output.json'
output_path = '/content/drive/MyDrive/Project/finetune data/11_30_filtered_combined_output.json'

with open(filter_words_path, 'r', encoding='utf-8') as file:
    filter_words = json.load(file)
# with open(combined_path, 'r', encoding='utf-8') as file:
#     combined_json = json.load(file)

########################
with open(original_combined_path, 'r', encoding='utf-8') as file:
    original_combined_json = json.load(file)
######################## Edited by 光齊

filtered_json = []
for prompt in original_combined_json:####### Filter the original combined data instead(filter those pairs with input="Story context before", not input="")
    if(contains_keyword(prompt["instruction"], filter_words)):
        filtered_json.append(prompt)
print(filtered_json[-1])
######################## Also add short_combined_output.json
# for prompt in combined_json:
#     filtered_json.append(prompt)
########################
print(filtered_json[-1])

# print(len(filtered_json))
with open(output_path, 'w', encoding='utf-8') as file:
    json.dump(filtered_json, file, indent=2, ensure_ascii=False)

print(len(original_combined_json))
print(len(filtered_json))

reduced_percentage = (len(original_combined_json) - len(filtered_json))/ len(original_combined_json) * 100
print(f"Reduced {reduced_percentage:.2f}%")

# reduced_percentage = (len(original_combined_json) - len(filtered_json) + len(combined_json))/ len(original_combined_json) * 100
# print(f"Reduced {reduced_percentage:.2f}%")

# print(data)
```
