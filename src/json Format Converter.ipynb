```python
import json

def convert_format(data):
    """
    Convert data from format A to format B.

    Format A:{'input':,'output':,'instruction':}
    Format B:[{'from':'human','value':},{,'from':'gpt','value':}]

    Args:
        data (dict): Data in format A.

    Returns:
        list: Data in format B.
    """
    instruction = data.get("instruction", "")
    input_text = data.get("input", "")
    output_text = data.get("output", "")

    format_b = {
        "conversations": [
            {"from": "human", "value": instruction + ("\n\n" + input_text if input_text else "")},
            {"from": "gpt", "value": output_text}
        ],
        "source": "gpt4o"#,"score": 4.0

    }

    return format_b


# File paths
input_path = '/content/drive/MyDrive/Project/finetune data/chatgpt.json'
output_path = '/content/drive/MyDrive/Project/finetune data/chatgpt_llama32ver.json'

# Read from input file
with open(input_path, 'r') as infile:
    data = json.load(infile)

# Check if the data is a list or a single object
if isinstance(data, list):
    converted_data = [convert_format(item) for item in data]
else:
    converted_data = [convert_format(data)]

# Write to output file
with open(output_path, 'w') as outfile:
    json.dump(converted_data, outfile, indent=4)

print(f"Conversion complete. Data written to {output_path}")
```
